```python
from Bio import SeqIO
from Bio.Seq import Seq
import numpy as np
import pandas as pd
import random
```

```python
meta=[]
sequence=[]
```

```python
seq = ('Motif_NonCLIP.txt')
```

```python
for seq_record in SeqIO.parse(seq, "fasta"):
    meta.append(str(seq_record.id))
    sequence.append(str(seq_record.seq))
```

```python
Fasta = pd.DataFrame(data ={'Meta':meta,'Sequence':sequence})
```

```python
Fasta
```

```python
Data = pd.read_csv("Motif_NonCLIP.csv")
```

```python
Data
```

```python
Result = pd.merge(Data, Fasta, left_index=True, right_index=True, how='outer')
```

```python
Result
```

```python
Result.drop(["Meta"], axis=1, inplace=True)
```

```python
Result["GeneLength"] = Result["txEnd"] - Result["txStart"]
Result["LastExonLength"] = Result["LastExonEnd"] - Result["LastExonStart"]
```

```python
Result
```

```python
Data = Result
```

```python
Data["Pre"] = Data["GeneLength"] - Data["LastExonLength"]
```

```python
Data
```

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
Sequence = Negative["Sequence"].tolist()
```

```python
New = []
for i in range(len(Sequence)):
    input_sequence = Sequence[i][::-1]
    input_sequence = Seq(input_sequence).complement()
    input_sequence = str(input_sequence)
    New.append(input_sequence)

New = np.array(New)
New = pd.DataFrame({"Sequence":New})
```

```python
Negative["Sequence"] = 0
Negative["Sequence"] = New["Sequence"]
```

```python
Negative
```

```python
Data = Positive.append(Negative)
Data = Data.reset_index(drop = True)
```

```python
Data
```

### Get sequence for LastExon

```python
Data["PreSequence"] = 0
Data["LastExonSequence"] = 0
```

```python
for i in range(len(Data)):
    Data.loc[i,("PreSequence")] = Data.loc[i,("Sequence")][:(Data.loc[i,("Pre")])]
    Data.loc[i,("LastExonSequence")] = Data.loc[i,("Sequence")][(Data.loc[i,("Pre")]):]
```

```python
Data
```

### Define m6A position in LastExon

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
Positive["m6APosition"] = Positive["m6AStart"] - Positive["LastExonStart"]
Negative["m6APosition"] = Negative["LastExonEnd"] - Negative["m6AEnd"]
```

```python
Data = Positive.append(Negative)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data['m6ASequence'] = 0
```

```python
for i in range(len(Data)):
    Data.loc[i,("m6ASequence")] = Data.loc[i,("LastExonSequence")][(Data.loc[i,("m6APosition")]-1):(Data.loc[i,("m6APosition")]+2)]
```

```python
Data
```

```python
Data = Data[Data["m6APosition"]> 50]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data = Data[(Data["LastExonLength"]-Data["m6APosition"])> 50]
Data = Data.reset_index(drop = True)
```

```python
Data
```

### Define m6A in pre-mRNA

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
Positive["POS"] = Positive["m6AStart"] - Positive["txStart"]
Negative["POS"] = Negative["txEnd"] - Negative["m6AEnd"]
```

```python
Data = Positive.append(Negative)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data["m6ASequence"] = Data["m6ASequence"].str.upper()
```

```python
Data["m6ASequence"].value_counts()
```

```python
GAC = Data[Data["m6ASequence"]=="GAC"]
GAC = GAC.sample(n=647, random_state=3)

AAC = Data[Data["m6ASequence"]=="AAC"]
AAC = AAC.sample(n=353, random_state=3)
```

```python
Data = GAC.append(AAC)
```

```python
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data["m6ASequence"].value_counts()
```

```python
for i in range(len(Data)):
    Data.loc[i,("Up")] = Data.loc[i,("LastExonSequence")][:(Data.loc[i,("m6APosition")]-50)]
    Data.loc[i,("Mid")] = Data.loc[i,("LastExonSequence")][(Data.loc[i,("m6APosition")]-50):(Data.loc[i,("m6APosition")]+51)]
    Data.loc[i,("Down")] = Data.loc[i,("LastExonSequence")][(Data.loc[i,("m6APosition")]+51):]
```
