```python
Phylop = Phylop[[0,1,2,3,4,5]]
```

```python
Phylop.columns = ["chrom","Start","End","Score","name","strand"]
```

```python
Phylop
```

```python
Phylop.drop_duplicates(subset=["chrom","Start","End","Score","name","strand"], keep="first", inplace=True)
```

```python
Name = Fasta["name"].tolist()
strand = Fasta["strand"].tolist()
```

```python
dat = pd.DataFrame(range(200),columns=["index"])
for i in range(len(Fasta)):
    df = pd.DataFrame(np.random.randn(201,3))
    df.columns = ["chrom", "name", "strand"]
    df["chrom"] = Fasta.loc[i,"chrom"]
    df["name"] = Fasta.loc[i,"name"]
    df["strand"] = Fasta.loc[i,"strand"]

    list = range((Fasta.loc[i,"m6AStart"]-100), (Fasta.loc[i,"m6AStart"]+101))
    Start = pd.DataFrame(list, columns=["Start"])
    df = pd.concat([df, Start], axis=1)
    df["End"] = df["Start"] + 1
    df["Start"] = df["Start"].astype(int)
    df["End"] = df["End"].astype(int)
    
    df = pd.merge(df, Phylop, how="left", on=["chrom","Start","End","strand","name"])
    
    if strand[i] == "+":
        df = df.reset_index(drop = False)
        df = df[["index","Score"]]
        df.rename(columns={"Score":Name[i]}, inplace = True) 
    if strand[i] == "-":
        df.sort_index(inplace=True, ascending=False)
        df = df.reset_index(drop = True)
        
        df = df.reset_index(drop = False)
        df = df[["index","Score"]]
        df.rename(columns={"Score":Name[i]}, inplace = True)
    
    dat = pd.merge(dat, df, how="outer", on=["index"])
    
```

```python
dat.drop(["index"], axis=1, inplace=True)
dat = dat.T
```

```python
dat.loc["col_sum"] = dat.sum()
dat = dat.T
```

```python
Value = dat[["col_sum"]]
```

```python
dat.drop(["col_sum"], axis=1, inplace=True)
dat = dat.T
```

```python
dat.loc["Number"] = dat.isnull().sum()
dat = dat.T
```

```python
dat["Number"] = len(Fasta) - dat["Number"]
```

```python
Number = dat[["Number"]]
```

```python
dat.drop(["Number"], axis=1, inplace=True)
dat["SEM"] = dat.sem(axis=1)
SEM = dat[["SEM"]]
```

```python
dat.drop(["SEM"], axis=1, inplace=True)
dat["Mean"] = dat.mean(axis=1)
Mean = dat[["Mean"]]
```

```python
SEM
```

```python
Mean
```

```python
High = pd.concat([Mean, SEM], axis=1)
```
