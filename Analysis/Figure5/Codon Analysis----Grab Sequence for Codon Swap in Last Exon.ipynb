## Grab sequence

```python
import numpy as np
import pandas as pd
```

```python
Data = pd.read_csv("/pod/2/ke-lab/LUOZ/m6AAI/UniqmRNA_GENCODE_VM7com_mm10_RefSeqpriority.txt", sep="\t", header=None)
```

```python
Data.drop([0,11,13,14,15], axis=1, inplace=True)
```

```python
Data
```

```python
Data[1] = Data[12]
```

```python
Data.drop([12], axis=1, inplace=True)
```

```python
Data
```

```python
Data.columns = ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]
```

```python
Data
```

### Select coding genes

```python
idx = (Data["cdsStart"] != Data["cdsEnd"])
```

```python
Data = Data.loc[idx, ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]]
```

```python
Data
```

### Keep standard chromosomes

```python
Data = Data[:22357]
```

### exonCount > 1

```python
Data = Data[Data.exonCount > 1]
```

```python
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data["LastExonStart"] = 0
Data["LastExonEnd"] = 0
```

```python
Data
```

```python
for i in range(len(Data)):
    if Data.loc[i, "strand"] == "+":
        Data.loc[i,("LastExonStart")] = Data.loc[i, "exonStarts"].split(",")[-2]
        Data.loc[i,("LastExonEnd")] = Data.loc[i, "exonEnds"].split(",")[-2]
    elif Data.loc[i, "strand"] == "-":
        Data.loc[i,("LastExonStart")] = Data.loc[i, "exonStarts"].split(",")[0]
        Data.loc[i,("LastExonEnd")] = Data.loc[i, "exonEnds"].split(",")[0]
```

```python
Data
```

```python
Data.drop(["exonCount", "exonStarts", "exonEnds"], axis=1, inplace=True)
```

```python
Data
```

### m6A sites

```python
Site = pd.read_csv("M6A_CodonSwap.csv")
```

```python
Site
```

```python
Site.rename(columns={"Start":"m6AStart","End":"m6AEnd"}, inplace = True)
```

```python
Overlap = pd.merge(Data, Site, how="inner", on=["chrom", "name", "strand"])
```

```python
Overlap
```

```python
Overlap["LastExonStart"] = Overlap["LastExonStart"].astype(int)
Overlap["LastExonEnd"] = Overlap["LastExonEnd"].astype(int)
```

```python
Positive = Overlap[Overlap["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
```

```python
Negative = Overlap[Overlap["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

### Select m6A sites in the Coding region of LastExon

```python
Positive = Positive[(Positive["m6AStart"] < Positive["cdsEnd"])&(Positive["m6AStart"] > Positive["LastExonStart"])]
```

```python
Positive
```

```python
Negative = Negative[(Negative["m6AStart"] > Negative["cdsStart"])&(Negative["m6AStart"] < Negative["LastExonEnd"])]
```

```python
Negative
```

```python
Data = Positive.append(Negative)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data.to_csv("CodonSwap.csv", index=0)
```

### Create Bed

```python
Bed = Data.iloc[:, [1, 3, 4]]
```

```python
Bed.to_csv("CodonSwap.bed", sep="\t", header=False, index=False)
```
