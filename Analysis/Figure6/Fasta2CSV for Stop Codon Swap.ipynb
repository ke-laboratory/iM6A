```python
from Bio import SeqIO
import numpy as np
import pandas as pd
```

```python
meta=[]
sequence=[]
```

```python
seq = ('StopCodon.txt')
```

```python
for seq_record in SeqIO.parse(seq, "fasta"):
    meta.append(str(seq_record.id))
    sequence.append(str(seq_record.seq))
```

```python
Fasta = pd.DataFrame(data ={'Meta':meta,'Sequence':sequence})
```

```python
Fasta
```

```python
Data = pd.read_csv("UniqmRNA_GENCODE_VM7com_mm10_RefSeqpriority_Stop_Codon.csv")
```

```python
Data
```

```python
Result = pd.merge(Data, Fasta, left_index=True, right_index=True, how='outer')
```

```python
Result
```

```python
Result.drop(["Meta"], axis=1, inplace=True)
```

```python
Result
```

```python
Result["Length"] = Result["txEnd"] - Result["txStart"]
```

```python
Result
```

```python
Result = Result.sort_values(by=["chrom","txStart"], axis=0, ascending=True)
Result = Result.reset_index(drop = True)
```

```python
Result.to_csv("Fasta_Stop_Codon.csv", index=0)
```
