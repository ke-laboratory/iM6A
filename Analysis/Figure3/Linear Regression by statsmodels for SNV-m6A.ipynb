```python
import statsmodels.api as sm
```

```python
Intercept = []
Slope = []
Std = []
R2 = []
R2adj = []
Pvalue = []
T = []
F = []
Fpvalue = []
AIC = []
BIC = []

for i in range(len(m6A)):
    Score = pd.DataFrame(np.array(SNP.iloc[i,2:62]), columns=["Score"])
    Value = pd.DataFrame(np.array(m6A.iloc[i,2:62]), columns=["Value"])
    data = pd.merge(Score, Value, left_index=True, right_index=True)
    
    x = np.asarray(data["Score"], dtype=float)
    y = np.asarray(data["Value"], dtype=float)
    x = sm.add_constant(x)
    model = sm.OLS(y, x).fit()
    
    r2 = model.rsquared
    adj = model.rsquared_adj
    
    intercept = model.params[0]
    slope = model.params[1]
    
    t = model.tvalues[0]
    pvalue = model.pvalues[0]
    std = model.bse[0]
    f = model.fvalue
    fpvalue = model.f_pvalue
    aic=model.aic
    bic= model.bic
    
    Slope.append(slope)
    Intercept.append(intercept)
    R2.append(r2)
    R2adj.append(adj)
    Pvalue.append(pvalue)
    Std.append(std)
    T.append(t)
    F.append(f)
    Fpvalue.append(fpvalue)
    AIC.append(aic)
    BIC.append(bic)
```

```python
List = [Slope, Intercept,T, R2, R2adj, Pvalue, Std, F, Fpvalue, AIC, BIC]
df = pd.DataFrame(List)
df = df.T
df.rename(columns={0:"Slope", 1:"Intercept",2:"T", 3:"R2", 4:"R2_adj", 5:"Pvalue", 6:"Std", 7:"F", 8:"Fpvalue", 9:"AIC", 10:"BIC"}, inplace = True)
df
```

```python
from Bio import SeqIO
from Bio.Seq import Seq
```

```python
meta=[]
sequence=[]
seq = ('m6A_QTL_SigSite.txt')
```

```python
for seq_record in SeqIO.parse(seq, "fasta"):
    meta.append(str(seq_record.id))
    sequence.append(str(seq_record.seq))
```

```python
Fasta = pd.DataFrame(data ={'Meta':meta,'Sequence':sequence})
Result = pd.merge(Data, Fasta, left_index=True, right_index=True, how='outer')
```

```python
Result.drop(["Meta"], axis=1, inplace=True)
Result["GeneLength"] = Result["txEnd"] - Result["txStart"]
```
